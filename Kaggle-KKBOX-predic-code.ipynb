### Check if there is any missing value in each column of te

```python
colNumber = len(te.columns)     
columnHeader = list(te.columns)
Result = {}

#run this for loop for 44 columns
for col in range (0, colNumber):   
    te_subset = te.iloc[ : , col]  #subset one column and all the rows
    series = pd.Series(te_subset)  #convert it into pd.Series object
    
    MissingValueCheck = series.isnull().values.any()  #If there is missing data, return TRUE
    
    Key = columnHeader[col]          #get the column name
    Result[Key] = MissingValueCheck  #store the checking result in a dictionary
```

```python
Result
```

```python
#filter out those columns with missing value
KeyList = Result.keys()
for KeyNumber in range (0,len(KeyList)):
    if Result[KeyList[KeyNumber]] == True:
        print KeyList[KeyNumber]
```

### Q23: create ROC curve and get the area-under-the-curve

```python
actual = y_test
predictions = p
```

```python
plt.title('ROC curve')
plt.plot(false_positive_rate, true_positive_rate, 'b', label = 'AIC = %0.2f' % roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.2])
plt.ylim([-0.1,1.2])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
